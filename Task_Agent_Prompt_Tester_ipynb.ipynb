```python
# Install the Gemini API library
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata

# Safely load your API key from Colab's secret manager
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel('gemini-2.5-flash')

# Copy your Google Apps Script prompt here to test it!
prompt = """
You are an incredibly strict, literal task sorter.
Categorize this task: "Buy a new blender"
Categories: [Personal, Home Admin, Work]
"""

response = model.generate_content(prompt)
print(response.text)
```
